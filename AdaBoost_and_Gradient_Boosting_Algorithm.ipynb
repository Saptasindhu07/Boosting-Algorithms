Extracting the Numeric and Categorical Features


```python
numeric_features= data.select_dtypes(include=np.number)
categorical_features=data.select_dtypes(exclude=np.number)
```

```python
discrete_features=[feature for feature in numeric_features if len(data[feature].unique())<25]
continuous_features=[feature for feature in numeric_features if feature not in discrete_features]
```

```python
discrete_features=data[discrete_features]
continuous_features=data[continuous_features]
```

#ENCODING THE CATEGORICAL FEATURE COLUMNS

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, StandardScaler
```

```python
from sklearn.compose import ColumnTransformer
```

```python
continuous_features_copy=continuous_features.copy()
standard_scaler=StandardScaler()
continuous_features=standard_scaler.fit_transform(continuous_features)
continuous_features=pd.DataFrame(continuous_features,columns=continuous_features_copy.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('ProdTaken',axis=1),df['ProdTaken'],test_size=0.2)
```

# **Training Our AdaBoost Model**

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score,precision_score,recall_score,f1_score
from scipy.stats import uniform, randint
```

#Hyperparameter Tuning

```python
rf_params={
    'max_depth':[5,8,15,None,10,12,15,17],
    "max_features":[2,5,7,8,"auto"],
    "min_samples_leaf":[1,3,5,6,8,10],
    "n_estimators":[100,200,300,400]
}
ada_boost_params={
    'estimator':[DecisionTreeClassifier(max_depth=1),DecisionTreeClassifier(max_depth=2),DecisionTreeClassifier(max_depth=3),DecisionTreeClassifier(max_depth=7),DecisionTreeClassifier(max_depth=12),DecisionTreeClassifier(max_depth=15)],
    'n_estimators':[50,70,90,100],
    'learning_rate':[0.2,0.5,1,2,3,5,7,9,10,12,13,14,15,16,17,19,20]
}
gradient_boost_params={
    'loss':['log_loss', 'exponential'],
    'criterion':['mae','mse','rmse'],
    'learning_rate': uniform(0.01, 0.29),  # Continuous: 0.01 to 0.3
    'n_estimators': randint(50, 500),     # Discrete: 50 to 500 trees
    'max_depth': randint(3, 20),          # Discrete: 3 to 10
    'min_samples_split': randint(2, 20),  # Discrete: 2 to 20
    'min_samples_leaf': randint(1, 20),   # Discrete: 1 to 20
    'subsample': uniform(0.5, 0.5),       # Continuous: 0.5 to 1.0
    'max_features': ['sqrt', 'log2',None],  # Options: sqrt, log2, or all features
    'ccp_alpha': uniform(0.0, 0.1),
}
randomvcModels=[
    ("RF",RandomForestClassifier(),rf_params),
    ("AdaBoost", AdaBoostClassifier(),ada_boost_params),
    ("Gradient Boosting", GradientBoostingClassifier(),gradient_boost_params)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_param={}

for name,model,params in randomvcModels:
  random=RandomizedSearchCV(estimator=model,param_distributions=params,n_iter=50,cv=5,verbose=2,n_jobs=-1,scoring='f1')


  random.fit(X_train,y_train)
  model_param[name]=random.best_params_
  print(f"{name}")
  print("-------------------------")
  print("Recall: ",recall_score(random.predict(X_test),y_test))
  print("Precision: ",precision_score(random.predict(X_test),y_test))
  print("Accuracy: ",accuracy_score(random.predict(X_test),y_test))
  print("Confusion Matrix"+"\n"+str(confusion_matrix(random.predict(X_test),y_test)))
  print("Classification Report"+"\n"+str(classification_report(random.predict(X_test),y_test)))
  print("F1 Score: ",f1_score(random.predict(X_test),y_test))

for model_name in model_param:
  print(f"Best Parameters for {model_name}: {model_param[model_name]}")
```
